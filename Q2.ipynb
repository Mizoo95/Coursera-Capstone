```python
import csv 
import pandas as pd 
import numpy as np 
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

### Scraping for needed data

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
html = urlopen(url)
soup = BeautifulSoup(html)
```

```python
headers = []
header_list = soup.find_all('th')
for col in header_list:
    headers.append(col.text)
headers = headers[:3]
```

```python
data = []
rows = soup.find_all('tr')
for row in rows:
    row_list = row.find_all('td')
    DataRow = []
    
    for cell in row_list:
        DataRow.append(cell.text)
    data.append(DataRow)
data = data[1:]
#print(data)
```

### Creating data frame

```python
df = pd.DataFrame(data)
col_drop = range(3,31)
df = df.drop(col_drop, axis=1)
df
```

```python
#Specifying Columns names
df.columns = headers
df
```

```python
#Dropping unrelated data

col_drop = range(287,293)
df = df.drop(col_drop, axis=0)
df.tail(10)
```

```python
#Edit columns names
df.rename(columns={'Neighbourhood\n': 'Neighbourhood'}, inplace=True)
```

```python
df.dtypes
```

```python
df['Neighbourhood'] = df.Neighbourhood.str.replace('\n','')
pd.options.mode.use_inf_as_na = True
```

```python
#Changing Not assigned data to NaN for simplicity
df['Borough'] = df.Borough.str.replace('Not assigned', 'NaN')
df['Neighbourhood'] = df.Neighbourhood.str.replace('Not assigned', 'NaN')
df.head(20)
```

#### Removing rows that contains NaN values in column Borough

```python
for i,j in enumerate(df['Borough']):
    if(j == 'NaN'):
        df.drop(index= i, axis=0, inplace=True)
```

```python
#remove duplicates from the dataframe df
df = df.drop_duplicates(keep='first')
df.head(50)
```

```python
#reset the data frame index for simplicity
df2 = df.reset_index()
df2.drop('index', axis=1,inplace=True)
df2.head(20)
```

#### Setting NaN values in 'Neighbourhood' column as tha same as 'Borough'

```python
for i,j in enumerate(df2.Neighbourhood):
    if(j == 'NaN'):
        df2.Neighbourhood[i] = df2.Borough[i]
        print(j)
```
